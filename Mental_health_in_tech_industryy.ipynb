```python
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins=20)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df['Gender'].value_counts().head(5).plot(kind='bar')
plt.title("Gender Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='treatment', data=df)
plt.title("Treatment Seeking")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='family_history', hue='treatment', data=df)
plt.title("Family History vs Treatment")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='work_interfere', data=df)
plt.title("Work Interference")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='benefits', data=df)
plt.title("Mental Health Benefits")
plt.show()
```
